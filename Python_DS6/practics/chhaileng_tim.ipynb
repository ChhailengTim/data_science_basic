<b>Exercise 6. Calculate the total income in this dataset.</b>

```python
#Convert unit price to float (No $)
prices = []
for value in order_list.unit_price:
    value_f = float(value[1: -1])
    prices.append(value_f)
order_list.unit_price = prices
order_list.head()
```

```python
order_list.head()
```

```python
order_list['total_price'] = order_list['unit_price'] * order_list['quantity']
total_price = order_list['total_price'].sum()
total_price


```

<b>Exercise 7. Find the average income for each order?</b>

```python
average_income = (order_list[['id', 'total_price']].groupby('id', as_index=False).agg(total_by_order=('total_price', 'sum')))
average_income['total_by_order'].mean()
```

<b>Exercise 8. How many different foods are ordered?</b>

```python
dif_food = order_list.groupby('food').size().reset_index(name='count')
dif_food
```

<b>Exercise 9. Display the food which has the highest total quantity sold, and its quantity.</b>

```python
highest_order = dif_food.sort_values(by='count', ascending=False).head(1)
highest_order
```

<b>Exercise 10. What is the most frequently ordered food item?</b>

```python
frequent_food = highest_order['food'].values[0]
most_frequent = order_list[order_list['food'] == frequent_food]
most_frequent
```
