```python
df['total_cost'] = df['Avg_Item_Price_USD'] * df['Avg_Spend_Per_Customer_USD']
df[['total_cost']].head()
```

```python
df['Upper'] = df['Country'].str.upper()    
df
```

```python
avg_cost_per_country = df.groupby('Country').agg(Avg_total_cost=('total_cost', 'mean')).head(5)
avg_cost_per_country
```

```python
sns.barplot(data=avg_cost_per_country, x='Country', y='Avg_total_cost', palette='Greens_d')

plt.title('Avg total cost per country)', fontsize=18)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Avg Cost Per Year (USD)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
